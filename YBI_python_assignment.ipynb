```python
import matplotlib.pyplot as plt
```

```python
x=[2,3,4,5,6,7,8,9]
y=[1,2,3,4,5,6,7,8]
```

```python
fig,ax = plt.subplots()
ax.plot(x,y)
plt.show()
```

```python
fig,ax = plt.subplots()
ax.bar(x,y)
plt.show()
```

```python
fig,ax = plt.subplots()
ax.barh(x,y)
plt.show()
```

```python
fig,ax = plt.subplots()
ax.scatter(x,y)
plt.show()
```

```python
fig,ax = plt.subplots()
ax.hist(x)
plt.show()
```

```python
fig,ax = plt.subplots()
ax.pie(x)
plt.show()
```

```python
fig,ax = plt.subplots()
ax.plot(x,y , color = 'red' , linestyle = '--', marker = 'o' , linewidth = 2)
ax.set_title('Sales numbers for FY 2022')
ax.set_xlabel('Months')
ax.set_ylabel('Sales Unit numbers')
plt.show()
```
